```python
import os
import sys
import cv2
```

```python
raw_video_dict = '../ROM_raw_videos'
raw_video_clip_dict = '../ROM_raw_videos_clips'
#dates = os.listdir(raw_video_dict)
#dates = ['Sep_12','Sep_13','Sep_14','Sep_17','Sep_18','Sep_19','Sep_20',
#         'Sep_21','Sep_22','Sep_23','Sep_24','Sep_25','Sep_26','Sep_27','Sep_28',
#         'Oct_02', 'Oct_03']
dates = ['Oct_05']
for date in dates:
    input_directory = os.path.join(raw_video_dict, date)#'../ROM_raw_videos/{}'.foramt(date)
    output_directory = os.path.join(raw_video_clip_dict, date)#'../ROM_raw_videos_clips/{}'.format(date)
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
    
    for filename in os.listdir(input_directory):
        camera = cv2.VideoCapture(os.path.join(input_directory,filename))
        total_frame_number = camera.get(cv2.CAP_PROP_FRAME_COUNT)
        fps = camera.get(cv2.CAP_PROP_FPS)
        duration = (total_frame_number/fps) # duration in second
        print('Video duration: {}s'.format(duration))
        if duration > 3000: 
            clip_number = 6 # 1 hour video, cut into 6 clips
        else: 
            clip_number = 3 # half hour video, cut into 3 clips

        clip_duration = round(duration/clip_number)
        name, _ = filename.split('.mp4')
        for i in range(clip_number):
            input_filename = os.path.join(input_directory, filename)
            output_filename = os.path.join(output_directory, '{}_{}.mp4'.format(name, i))

            print(output_filename)
            os.system('cd {}'.format(os.path.abspath(output_directory)))
            code = os.system('ffmpeg -ss {} -i {}  -t {} -c copy {}'.format(i*clip_duration,
                                                                            input_filename, 
                                                                            clip_duration, 
                                                                            output_filename))
            print(code)
```
