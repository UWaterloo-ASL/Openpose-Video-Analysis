```python
import os
import glob
```

## Camera2

```python
#output_dir = '../ROM_raw_videos_clips_processed'
output_dir = '../ROM_raw_videos_clips_processed_camera2'
output_sub_dir = os.listdir(output_dir)
output_sub_dir
```

```python
copy_dir = './occupancy_estimation_camera2'
```

```python
for date in output_sub_dir:
    copy_sub_dir = os.path.join(copy_dir, date)
    if not os.path.exists(copy_sub_dir):
        os.makedirs(copy_sub_dir)
        
    print('On {}:'.format(date))
    files = glob.glob(os.path.join(output_dir, date, '*.csv'))
    for f in files:
        code = os.system('cp {} {}'.format(f, copy_sub_dir))
        print('{}: {}'.format(f, code))
```
